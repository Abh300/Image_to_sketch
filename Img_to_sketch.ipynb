```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#Add image which you want to create a sketch of along with image name
your_img = cv2.imread("pets.jpg")
your_img = cv2.cvtColor(your_img,cv2.COLOR_BGR2RGB)
plt.imshow(your_img)
```

```python
gray_img = cv2.cvtColor(your_img,cv2.COLOR_RGB2GRAY)
red_ch = your_img[:,:,0]
green_ch = your_img[:,:,1]
plt.imshow(green_ch,cmap='gray')
```

```python
img_invert = cv2.bitwise_not(gray_img)
img_inv2 = cv2.bitwise_not(green_ch)
plt.imshow(img_invert)
```

```python
smooth_img = cv2.GaussianBlur(img_invert,(101,101), sigmaX=0,sigmaY=0)
plt.imshow(smooth_img)
```

```python
def dodgeV2(x,y):
    return cv2.divide(x,245 - y ,scale=256)

final_image = dodgeV2(gray_img,smooth_img)
plt.imshow(final_image,cmap='gray')
```
